```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.countplot(x='age', data=df)
plt.title("Age Distribution")
plt.xlabel("Age (lower bound of range)")
plt.ylabel("Count")
plt.show()
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f"\n{col} value counts:")
    print(df[col].value_counts())
```

```python
df.groupby('age')['time_in_hospital'].mean().plot(kind='bar', figsize=(8, 4), title="Average Time in Hospital by Age")
plt.ylabel("Avg. Days")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='age', y='number_diagnoses', data=df)
plt.title("Number of Diagnoses by Age")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
readmission_analysis = df.groupby('readmitted')['number_diagnoses'].mean()
print(readmission_analysis)
```

```python
from sklearn.cluster import KMeans

# Select relevant columns for clustering
df_clustering = df[['time_in_hospital', 'number_diagnoses', 'age']]

# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_clustering_scaled = scaler.fit_transform(df_clustering)

# Fit KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_clustering_scaled)

# Visualize the clusters
sns.scatterplot(x=df['time_in_hospital'], y=df['number_diagnoses'], hue=df['cluster'], palette='viridis')
plt.title('Patient Clusters Based on Time in Hospital and Number of Diagnoses')
plt.show()
```

```python
# Pairplot to show relationships between multiple variables
sns.pairplot(df[['time_in_hospital', 'number_diagnoses', 'age']])
plt.show()
```

```python
# Gender distribution
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution of Patients')
plt.show()

# Age distribution 
sns.countplot(data=df, x='age')
plt.title('Age Distribution of Patients')
plt.xticks(rotation=45)
plt.show()

# Race distribution
sns.countplot(data=df, x='race')
plt.title('Race Distribution of Patients')
plt.xticks(rotation=45)
plt.show()
```

```python
# Time in hospital vs. readmission status
sns.boxplot(x='readmitted', y='time_in_hospital', data=df)
plt.title('Time in Hospital vs. Readmission')
plt.show()

# Average time in hospital for readmitted patients
readmission_avg = df.groupby('readmitted')['time_in_hospital'].mean()
print(readmission_avg)
```

```python
# Most frequent primary diagnosis (diag_1)
top_diag_1 = df['diag_1'].value_counts().head(10)
print(top_diag_1)

# Diagnoses related to readmission
diag_readmission = df.groupby('diag_1')['readmitted'].value_counts().unstack().fillna(0)
print(diag_readmission)
```

```python
# Correlation matrix
corr_matrix = df[['time_in_hospital', 'number_diagnoses', 'age', 'num_lab_procedures']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# Gender vs. Number of Diagnoses
sns.boxplot(x='gender', y='number_diagnoses', data=df)
plt.title('Gender vs. Number of Diagnoses')
plt.show()

# Calculate average number of diagnoses by gender
avg_diag_by_gender = df.groupby('gender')['number_diagnoses'].mean()
print(avg_diag_by_gender)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select relevant features for clustering
features = df[['time_in_hospital', 'number_diagnoses']]

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Fit the K-Means model (3 clusters for this example)
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(features_scaled)

# Visualize clusters
sns.scatterplot(data=df, x='time_in_hospital', y='number_diagnoses', hue='cluster', palette='viridis')
plt.title('Patient Clusters Based on Time in Hospital and Number of Diagnoses')
plt.show()
```

```python
# Calculate the mean time in hospital and number of diagnoses for each cluster
cluster_analysis = df.groupby('cluster')[['time_in_hospital', 'number_diagnoses']].mean()
print(cluster_analysis)
```
